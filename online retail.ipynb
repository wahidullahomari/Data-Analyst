## EDA 

```python
## negative numbers from quantity and unitprice 
returns = df[df['Quantity'] < 0]
```

```python
df = df[(df['Quantity'] > 0 ) & (df['UnitPrice'] > 0 )]
```

```python
df
```

```python
## create total price of each transaction

df['totalprice'] = df['Quantity'] * df['UnitPrice']
```

```python
df
```

```python
df['Description'].value_counts().head(10)
```

```python
## total spending per customers

customer_spending = df.groupby('CustomerID')['totalprice'].sum().sort_values(ascending =False)	
```

```python
customer_spending.head(10) 
```

```python
## best products 

top_product = df.groupby('Description')['Quantity'].sum().sort_values(ascending = False)
```

```python
top_product.head(10) 
```

```python
df['Month'] = df['InvoiceDate'].dt.to_period('M')
```

```python
## Monthly Revenue 

monthly_revenue = df.groupby('Month')['totalprice'].sum()
monthly_revenue
```

## Visualization

```python
import matplotlib 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12,6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.figure(figsize=(10, 6))
plt.title('monthly_revenue')
plt.xlabel('Month')
plt.ylabel('totalprice')
plt.xticks(rotation=75)
sns.barplot(monthly_revenue, color='skyblue', edgecolor='black');
```

```python
# Top 10 products by quantity sold
top_products = data.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_products.plot(kind='barh', color='salmon', edgecolor='black')
plt.title('Top 10 Products Sold', fontsize=16)
plt.xlabel('Total Quantity Sold', fontsize=12)
plt.ylabel('Product Description', fontsize=12)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
